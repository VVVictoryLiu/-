# 股票获取与加载SAR买卖指标

```python
def generate_tech_indicators(dataframe):

        copy = dataframe[dataframe['tradestatus']==1].copy()
        sar,channel = tech.SAR(copy['high'],copy['low'],copy['close'])
        sar_trigger = tech.sar_trigger(channel)
        sar_trigger.index = copy.index
        sar.index = copy.index
        channel.index = copy.index
        copy['sar_trigger'] = sar_trigger
        copy['sar'] = sar
        copy['channel'] = channel

        return copy
```

```python
for i in list_put[:500]:
    sql = 'select date,code,open,high,low,close,preclose,amount,turn,tradestatus from stock_zz500 where code="{}"'.format(i)
    data = pd.read_sql(sql=sql,con=conn,index_col='date').sort_index()
    temp = generate_tech_indicators(data)[['sar_trigger','sar','close','high']]
    select_range = temp[temp.index>'2020-07-24'][['sar_trigger','sar','close','high']]
    if len(select_range)==0:
        continue
    
    select_range['gap'] = select_range['sar'] - select_range['high']
    
    if (select_range['gap'][-1]>0)&(select_range['gap'][-1]<select_range['high'][-1]*0.02):
        print('********************注意即将突破'+'*******************')
        print(i)
        print(select_range)
    
print('$'*10)
                     
    
    
```
